```python
#convert time_diff to minutes
df2['time_diff_minutes'] = df2['time_diff'].dt.total_seconds() / 60
display(df2['time_diff_minutes'])
```

Using shift()<br>--to calculate differences over time:

```python
#shift sentiment_label to match time_diff
df2['prev_sentiment'] = df2['sentiment_label'].shift()
```

Using groupby()<br>--to split specific criteria and combine the results:

```python
#group by previous sentiment and calculate average time difference
avg_time_by_sentiment = df2.groupby('prev_sentiment')['time_diff_minutes'].mean().dropna().round(2)
#print(avg_time_by_sentiment)
```

Time between Negative Messages:

```python
negative = df2[df2['sentiment_label'] == 'Negative'].index.tolist()

for i in range(1, len(negative)):
    idx_current = negative[i]
    idx_prev = negative[i - 1]

    t_current = df.loc[idx_current, 'timestamp']
    t_prev = df.loc[idx_prev, 'timestamp']

    elapsed_minutes_negative = (t_current - t_prev).total_seconds() / 60
    print(f"Time between Negative at row {idx_prev} and {idx_current}: {elapsed_minutes_negative:.2f} minutes")
```

Can go Futher to Convert Minutes to Hours and Minutes:<br>Defining a Function

```python
def convert_minutes_to_hours_and_minutes(minutes):
    hours = int(minutes // 60)
    remaining_minutes = int(minutes % 60)
    return hours, remaining_minutes

converted_minutes = 1755.00
converted_time = convert_minutes_to_hours_and_minutes(converted_minutes)
print(f"{converted_minutes} minutes is equal to {converted_time[0]} hours and {converted_time[1]} minutes.")
```

Can go even Further and Convert to Days:<br>Defining a Function

```python
def convert_hours_minutes_to_days(hours, minutes):
  total_minutes = hours * 60 + minutes
  days = total_minutes / (24 * 60)
  return days

hours_input = 30 # using 30 you get one whole day and .25, or 1/4 of 24 hours (1 day) which equals .25 (24/6 = 4 and is equivalent to 25% of 100% of a day) #this value can be changed
print(hours_input, "hours input")
minutes_input = 0 #this value can be changed
print(minutes_input, "minutes input")
days_output = convert_hours_minutes_to_days(hours_input, minutes_input)
print(days_output, "when hours and minutes converted to days")
print(f"{hours_input} hours and {minutes_input} minutes is equal to {days_output:.2f} days.")
#If using the above as input 20 for hours and 15 for minutes you get 1.22. (one whole day [24 hours] and .22 of a day)
```
